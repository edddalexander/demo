```python
def get_daily_data(coin_id, date):
  
  

  url = f"{base_url}/{coin_id}/history/daily?date={date}"
  headers = {'X-Cg-Apikey': api_key}

  response = requests.get(url, headers=headers)
  if response.status_code == 200:
    return response.csv()
  else:
    print(f"Error fetching data for {coin_id} on {date}: {response.status_code}")
    return None

```

```python
def download_year_data(coin_id, data_format="csv", output_file="data.csv"):
  

    today = datetime.today()
    one_year_ago = today - timedelta(days=365)  # Adjust for leap years if needed

  
    data = []
    for date in (one_year_ago + timedelta(days=i) for i in range(365)):
        daily_data = get_daily_data(coin_id, date.strftime('%Y-%m-%d'))
        if daily_data:
          data.append(daily_data)
          time.sleep(1)  # Introduce a delay to avoid exceeding rate limits

  # Process and store data based on format
    if data_format == "csv":
        with open(output_file, 'w') as f:
      # Write CSV headers (adjust based on actual data structure)
          f.write("date,price_usd,market_cap\n")
          for day in data:
        # Extract relevant data and write to CSV
            f.write(f"{day['date']},{day['market_data']['current_price']['usd']},{day['market_data']['market_cap']['usd']}\n"
    else:
        print(f"Unsupported data format: {data_format}")
```
